```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# 데이터 생성
n_sample = 300
n_clusters = 3
```

```python
x, y = make_blobs(n_samples=n_sample, centers=n_clusters,random_state=42)
```

```python
# 그래프
plt.scatter(x[:,0], x[:,1],c='gray',label='original data')
plt.title('original data')
plt.xlabel('f1')
plt.ylabel('f2')
```

```python
# Kmeans 객체 생성 -> 학습
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# 학습
cluster_labels = kmeans.fit_predict(x)
```

```python
# 군집화 결과 시각화
# plt.scatter(x[:,0],x[:,1], c=cluster_labels, cmap='')
```

```python
# 최적의 k 찾기
result = []

for k in range(1,10):
    kmeans_temp = KMeans(n_clusters=k, random_state=42)
    kmeans_temp.fit_predict(x)
    result.append(kmeans_temp.inertia_)

plt.plot(range(1,10),result,marker='o')
plt.title('result')

plt.xlabel('k')
plt.ylabel('result')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
# 실루엣 스코어
sil_avg = silhouette_score(x, cluster_labels)
print(sil_avg)
```

```python
# 시각화를 위한 점수 계산
sil_values = silhouette_samples(x, cluster_labels)
print(sil_values)
```
