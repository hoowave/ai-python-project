```python
from domain.student_score_analysis import StudentScoreAnalysis
from service.plt_service import PltService

import numpy as np
import pandas as pd

############### Applcation Main ###############
# .venv 파일은 제외하고 업로드 하였습니다!!
# .venv 파일은 제외하고 업로드 하였습니다!!
# .venv 파일은 제외하고 업로드 하였습니다!!
def main():
    students = []                                                       # 학생 리스트 선언
    numbers = np.random.choice(range(1, 21), size=20, replace=False)    # 중복없이 학생번호 20번 생성
    for number in numbers:                                              # 20번 반복
        name = f'학생{number}'                                           # 이름 = 학생n
        math = np.random.randint(50, 101)                               # 수학점수 = 50~100
        eng = np.random.randint(50, 101)                                # 영어점수 = 50~100
        science = np.random.randint(50, 101)                            # 과학점수 = 50~100
        students.append(StudentScoreAnalysis(name,math,eng,science))    # 이름과 각 과목 점수로 학생 인스턴스 생성 및 학생리스트 추가 

    df = pd.DataFrame([s.to_dict() for s in students])     # 데이터프레임 생성
    pltService = PltService(df)     # PltService 생성 및 StudentScoreAnalysis DataFrame 초기화
    pltService.run()                # 그래프 생성

# 애플리케이션 진입점
main()
```
