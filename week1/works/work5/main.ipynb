```python
from domain.student_score_analysis import StudentScoreAnalysis
from service.plt_service import PltService

import numpy as np
import pandas as pd

############### Applcation Main ###############
# .venv    !!
# .venv    !!
# .venv    !!
def main():
    students = []                                                       #   
    numbers = np.random.choice(range(1, 21), size=20, replace=False)    #   20 
    for number in numbers:                                              # 20 
        name = f'{number}'                                           #  = n
        math = np.random.randint(50, 101)                               #  = 50~100
        eng = np.random.randint(50, 101)                                #  = 50~100
        science = np.random.randint(50, 101)                            #  = 50~100
        students.append(StudentScoreAnalysis(name,math,eng,science))    #           

    df = pd.DataFrame([s.to_dict() for s in students])     #  
    pltService = PltService(df)     # PltService   StudentScoreAnalysis DataFrame 
    pltService.run()                #  

#  
main()
```
