```python
from domain.customer_sales_analysis import CustomerSalesAnalysis
from service.plt_service import PltService

import numpy as np
import pandas as pd

############### Applcation Main ###############
# .venv 파일은 제외하고 업로드 하였습니다!!
# .venv 파일은 제외하고 업로드 하였습니다!!
# .venv 파일은 제외하고 업로드 하였습니다!!
def main():
    customers = []                                                        # 고객 리스트 선언
    productList = ['노트북', '스마트폰', '이어폰', '키보드', '마우스', '인형']  # 중복없이 고객번호 10번 생성
    numbers = np.random.choice(range(1, 11), size=10, replace=False)      # 중복없이 고객번호 10번 생성
    for number in numbers:                                                # 10번 반복
        name = f'고객_{number}'
        # 2024년 내 랜덤 날짜
        purchDate = pd.Timestamp('2024-01-01') + pd.to_timedelta(np.random.randint(0, 365), unit='D')
        # 상품명 한개으로 가져오기
        prodName = np.random.choice(productList)
        # 수량: 1~5개 랜덤
        amount = np.random.randint(1, 6)
        # 단가: 10,000 ~ 100,000원 단위 랜덤, 최소 천원단위로 설정했음
        price = np.random.randint(10, 101) * 1000
        # 고객 인스턴스 생성 및 고객 리스트 추가 
        customers.append(CustomerSalesAnalysis(name, purchDate, prodName, amount, price))

    df = pd.DataFrame([c.to_dict() for c in customers])         # 데이터프레임 생성
    pltService = PltService(df)     # PltService 생성 및 CustomerSalesAnalysis DataFrame 초기화
    pltService.run()                # 그래프 생성
    
# 애플리케이션 진입점
main()
```
