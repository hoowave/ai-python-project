```python
import matplotlib.pyplot as plt
```

```python
# 데이터
x = [1,2,3,4,5]
y = [2,3,5,7,13]
```

```python
# 그래프 선택->생성
plt.plot(x, y, 
         marker='o',
         linestyle='-',
         color='b',
         label='Numbers'
         )

# 그래프 설정
plt.title('line Graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# 그래프 출력
plt.grid(True)
plt.show()
```

```python
# 데이터
ctgy = ['a', 'b', 'c', 'd', 'e'] # axis X
val = [5, 7, 3, 9, 6] # axis Y

# 그래프 생성
#plt.bar(ctgy, val,
#        color='skyblue',
#        edgecolor='black',
#        )

plt.barh(ctgy, val,
        color='skyblue',
        edgecolor='black',
        )

# 설정
plt.title('Bar Chart')
plt.xlabel('Ctgy')
plt.ylabel('val')

plt.show()
```

```python
import numpy as np
```

```python
# data 준비
data = np.random.randn(1000)

# 그래프 생성
plt.hist(data,
         bins=20,
         color='orange',
         edgecolor='black'
         )

plt.title('Histogram')
plt.xlabel('value range')
plt.ylabel('Count')

plt.show()
```

```python
# 박스플롯
import seaborn as sns

data = np.random.rand(10, 10)
mask = np.triu(np.ones_like(data,dtype=bool))

sns.heatmap(data, annot=True, mask=mask)
plt.title('HeatMap')
plt.show()
```

```python
# 박스 플롯
tips = sns.load_dataset('tips')

# 그래프 생성
sns.boxplot(data=tips,
            palette='Set2',
            x='day',
            y='total_bill'
            )
plt.title('BoxPlot')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='muted', split=True)

plt.title('Chart')
plt.show()
```
